```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

nyc_zip = "Resources/NYC Zipcodes 2020.csv"

rolling_data_queen = "Resources/rollingsales_queens.xls"
```

```python
nyc_zip_data = pd.read_csv(nyc_zip)

nyc_zip_data.head()
```

```python
long_island_zip = [11101, 11102, 11103, 11104, 11105, 11106, 11109, 11120]


zip_code_df = nyc_zip_data.loc[nyc_zip_data["ZIP Codes"].isin(long_island_zip)]

print(zip_code_df.shape)
zip_code_df.head(20)
```

```python
queen_sale_data = pd.read_excel(rolling_data_queen,header=4)

queen_sale_data.shape
```

```python
queen_sale_data.head()
```

```python
queen_sale_data.columns
```

```python
long_island_sale_data_df = queen_sale_data.loc[queen_sale_data["ZIP CODE"].isin(long_island_zip)]

long_island_sale_data_df.head()
```

```python
building_class = long_island_sale_data_df["BUILDING CLASS AT TIME OF SALE"].value_counts()

building_class
```

```python
x_axis = long_island_sale_data_df["SALE DATE"]
y_axis = long_island_sale_data_df["SALE PRICE"]

plt.scatter(x_axis, y_axis, edgecolors = "black", facecolors = "blue")
plt.title("Property Sales in Long Island City")
plt.xlabel("Date")
plt.ylabel("Sale Price")

plt.show()
```
